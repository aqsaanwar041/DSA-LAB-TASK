```python
import matplotlib.pyplot as plt
import numpy as np
import math
n_values = np.arange(1, 1001)
def constant(n):
    return [1 for _ in n]

def log_n(n):
    return [math.log2(i) for i in n]

def linear(n):
    return [i for i in n]

def n_log_n(n):
    return [i * math.log2(i) for i in n]

def quadratic(n):
    return [i**2 for i in n]

def exponential(n):
    return [2**i if i < 20 else None for i in n]  
y_constant = constant(n_values)
y_log = log_n(n_values)
y_linear = linear(n_values)
y_nlogn = n_log_n(n_values)
y_quadratic = quadratic(n_values)
y_exponential = exponential(n_values[:20])  
# Plot
plt.figure(figsize=(12, 8))

plt.plot(n_values, y_constant, label='O(1) - Constant')
plt.plot(n_values, y_log, label='O(log n) - Logarithmic')
plt.plot(n_values, y_linear, label='O(n) - Linear')
plt.plot(n_values, y_nlogn, label='O(n log n)')
plt.plot(n_values, y_quadratic, label='O(n^2) - Quadratic')
plt.plot(n_values[:20], y_exponential, label='O(2^n) - Exponential')

plt.xlabel('Input size (n)')
plt.ylabel('Operations')
plt.title('Time Complexity Growth')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

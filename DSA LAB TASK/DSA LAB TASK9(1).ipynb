```python
def bubble_sort(arr):
    a = arr.copy()
    n = len(a)
    for i in range(n):
        for j in range(0, n-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    return a

def selection_sort(arr):
    a = arr.copy()
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a

def insertion_sort(arr):
    a = arr.copy()
    for i in range(1, len(a)):
        key = a[i]
        j = i-1
        while j >= 0 and key < a[j]:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key
    return a
```

```python
arr = [64, 25, 12, 22, 11]
print("Bubble Sort:", bubble_sort(arr))
print("Selection Sort:", selection_sort(arr))
print("Insertion Sort:", insertion_sort(arr))
```

```python
import time
import matplotlib.pyplot as plt
import random

sizes = [100, 200, 300, 400, 500]
bubble_times, selection_times, insertion_times = [], [], []

for size in sizes:
    test_arr = [random.randint(1, 1000) for _ in range(size)]

    start = time.time()
    bubble_sort(test_arr)
    bubble_times.append(time.time() - start)

    start = time.time()
    selection_sort(test_arr)
    selection_times.append(time.time() - start)

    start = time.time()
    insertion_sort(test_arr)
    insertion_times.append(time.time() - start)

plt.plot(sizes, bubble_times, label='Bubble Sort')
plt.plot(sizes, selection_times, label='Selection Sort')
plt.plot(sizes, insertion_times, label='Insertion Sort')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (s)')
plt.title('Sorting Algorithm Time Complexity')
plt.legend()
plt.grid(True)
plt.show()
```
